# Algorithm to process data to obtain quasi-stiffness of ankle.

```python
## Loading all necessary libraries
import pandas as pd
from bokeh.plotting import output_notebook, show, figure, curdoc
import warnings
from bokeh.layouts import row, widgetbox, column
from bokeh.models import Select
from bokeh.palettes import Spectral
from bokeh.plotting import curdoc, figure
output_notebook()
warnings.filterwarnings('ignore')
```

First of all, we're going to process the data as a csv file, the data was taken from the sto files generated by Opensim(r)  

```python
### converting a Storage opensim file into csv file
# Location dirs
ResultsDir = "/home/nikorose/Dropbox/Gait_Analysis_data/Models/Gait2354_Simbody"
ResultsDir2392 = "/home/nikorose/Dropbox/Gait_Analysis_data/Models/Gait2392_Simbody"
ResultsDirHenry = "/home/nikorose/Dropbox/Gait_Analysis_data/Prueba TTA/Corregido"
IKfile = ResultsDir+ "/IKresults.csv"
IK2392 = ResultsDir2392+'/subject01_walk1_ik.csv'
IKHenry = ResultsDirHenry+'/HenryAPIIK.csv'

##Model 2392 data
RRA2392 = ResultsDir2392+"/ResultsRRA/subject01_walk1_RRA_Kinematics_q.csv"
RRAActuatorFile2392 = ResultsDir+"/ResultsRRA/subject01_walk1_RRA_Actuation_force.csv"
IK2392csv = pd.read_csv(IK2392, header='infer', sep = r'\s*', engine='python')
RRA2392csv = pd.read_csv(RRA2392, header='infer', sep=r'\s*', engine='python')
RRAforcecsv = pd.read_csv(RRAActuatorFile2392, header='infer', sep=r'\s*', engine='python')

## Model 2354 data
RRA2354 = ResultsDir+"/ResultsRRA/subject01_walk1_RRA_Kinematics_q.csv"
IKcsv = pd.read_csv(IKfile, header='infer', sep= r"\s*", engine='python')
RRAActuatorFile = ResultsDir+"/ResultsRRA/subject01_walk1_RRA_Actuation_force.csv"
RRAforcecsv = pd.read_csv(RRAActuatorFile, header='infer', sep=r'\s*')
RRA2354csv = pd.read_csv(RRA2354, header='infer', sep=r'\s*', engine='python')

## Henry 2392 data
RRAhenry = ResultsDirHenry+"/ResultsRRA/Henry_walk1_RRA_Kinematics_q.csv"
IKhenry_csv = pd.read_csv(IKHenry, header='infer', sep= r"\s*", engine='python')
RRAhenry_ActuatorFile = ResultsDirHenry+"/ResultsRRA/Henry_walk1_RRA_Actuation_force.csv"
RRAhenry_forcecsv = pd.read_csv(RRAhenry_ActuatorFile, header='infer', sep=r'\s*')
RRA_henrycsv = pd.read_csv(RRAhenry, header='infer', sep=r'\s*', engine='python')
```

Now, We're going to compare the accuracy of the RRA algorithm to validate if the data is providing concordance. In order to do the checking, we compared the results of the inverse kinematics (IK) with literature and RRA algorithm in the following way: 

```python
cuadratura = 700
p1 = figure(x_axis_label='Time', y_axis_label='Right Ankle Angle',plot_width=cuadratura, plot_height=cuadratura)
p1.line(x = IKcsv['time'],y=IKcsv['ankle_angle_r'], color = "red", line_width = 2, legend = 'Inverse Kinematic 2354')
p1.line(x = RRA2354csv['time'],y=RRA2354csv['ankle_angle_r'], color = "blue", line_width = 2, legend = 'RRA Kinematics 2354')
minline = RRA2354csv.loc[RRA2354csv['ankle_angle_r'].idxmin()]
p1.ray(x=0.60, y=minline['ankle_angle_r'], length=10, angle=1.57079633, color='gray',line_dash="4 4",)


p2 = figure(x_axis_label='Time', y_axis_label='Right Ankle Angle',plot_width=cuadratura, plot_height=cuadratura)
p2.line(x = IK2392csv['time'],y=IK2392csv['ankle_angle_r'], color = "red", line_width = 2, legend = 'Inverse Kinematic 2392')
p2.line(x = RRA2392csv['time'],y=RRA2392csv['ankle_angle_r'], color = "blue", line_width = 2, legend = 'RRA Kinematics 2392')
minline2 = RRA_henrycsv.loc[RRA2392csv['ankle_angle_r'].idxmin()]
p2.ray(x=0.633, y=minline2['ankle_angle_r'], length=10, angle=1.57079633, color='gray',line_dash="4 4",)
p2.ray(x=1.867, y=minline2['ankle_angle_r'], length=10, angle=1.57079633, color='gray',line_dash="4 4",)

p3 = figure(x_axis_label='Time (s)', y_axis_label='Right Ankle Angle (deg)',plot_width=cuadratura, plot_height=cuadratura)
p3.line(x = IKhenry_csv['time'],y=IKhenry_csv['ankle_angle_r'], color = "red", line_width = 2, legend = 'Inverse Kinematic Henry')
p3.line(x = RRA_henrycsv['time'],y=RRA_henrycsv['ankle_angle_r'], color = "blue", line_width = 2, legend = 'RRA Kinematics Henry')
minline3 = RRA_henrycsv.loc[RRA_henrycsv['ankle_angle_r'].idxmin()]
p3.ray(x=3.26+.63, y=minline3['ankle_angle_r'], length=10, angle=1.57079633, color='gray',line_dash="4 4",)
p3.ray(x=4.43+.63, y=minline3['ankle_angle_r'], length=10, angle=1.57079633, color='gray',line_dash="4 4",)


show(column(p1,p2,p3),notebook_handle=True)
```

As the results were concordant, we'll proceed to calculate quasi-stiffness of ankle primarily with the RRA algorithm made by Opensim project.
