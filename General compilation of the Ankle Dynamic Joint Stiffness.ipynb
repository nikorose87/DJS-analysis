## 1.1 Introduction

Most of the researchers, who attempt to design the proper ankle-foot prosthesis, must study the biomechanics of the ankle joint in order to obtain the dynamic conditions of that part of the body, synthesizing this mechanical behavior in the angle-moment chart or also known as quasi-stiffness slope.  In addition, with the aim of mimicking the Dynamic Joint Stiffness (DJS) to be generated, most of the studies that relate the ankle quasi-stiffness tried to linearized each subphase of the human gait in consideration of getting the constant elastic coefficient of the material and/or mechanism that will return the energy.


In general, the DJS of the ankle will depend on the speed, weight, and height of the subject. However, abnormalities in gait affect the regular behavior of the DJS. With the rise of the study of this field, various investigations were done at particular circumstances and the need of establishing patterns according to specific parameters in gait have arrived.

First of all, we did a systematic search in the literature, research center websites and clinical databases. After founding the possible and reliable data to be analyzed, we selected some candidates (papers or databases) with the purpose of figuring out concordances in the results of each one ---in the case of having same inputs--- or to establishing differences according to the biomechanical parameters determined on each case of study.

Second of all, pure data of the selected candidates were extracted and filtered so that insights about the dynamic behavior of the ankle could describe unseen aspects of different types of gait.

Due to the above, we tried to identify better linear approximations at different instances of the ankle DJS and compared with non-linear approximations of the quasi-stiffness in terms of statistics. Moreover, with those similarities, we figure out the relationship between the ankle-moment and ankle-angle around all the loop, e.g. establishing gait sub-phases, work involved on the DJS, differences in gait parameters as gait speed, gender, age, different intentions, etc.

Also, this work attempts to serve as a meta-analysis of different literature data in order to describe the DJS at different conditions or gait intentions, this information could help to prosthesis designers so that tuning their active devices on each intention or type of gait.

### 1.3.1 Obtaining the work produced through the integration of the area under a curve with the Simpson's Rule

A piece of code calculates the work along the x-axis through the Simpon's Rule. Mathematically this rule is represented as: 
\begin{equation}
Area=\int_{a}^{b}f(x)dx\thickapprox\frac{\triangle x}{3}[y_{0}+4\sum odds+2\sum even+y_{n}]
\end{equation}

### 1.3.4 Ordinary least squares

As a linear regression technique, it fits a linear model with coefficients $w=(w_{1},...,w_{p})$ in order to minimize the residual sum of the squares between the observed responses in the dataset, and the response predicted by the linear approximation, in which this problem is represented mathematically as: $$\underset{w}{\min}||Xw-y||_{2}^{2}$$

Particularly, it is needed as many equations as the number of constants in the equation, formally called normal equations. In the straight line there are two constant so we require two normal equation:

Normal equation for a: $\sum Y=na+b\sum X$

Normal equation for b: $\sum XY=a\sum X+b\sum X^{2}$

The direct formula of finding a and b is written as: $$b=\frac{\sum XY-\frac{(\sum X)(\sum Y)}{n}}{\sum X^{2}-\frac{(\sum X)^{2}}{n}},\quad a=\bar{Y}-b\bar{X}$$

### 1.3.5 Ridge Regression

In order to avoid overfitting in our linearization, it is imposed a penalty to the ordinary least square method as it is presented here: $$\underset{w}{\min}||Xw-y||_{2}^{2}+\alpha||w||_{2}^{2}$$
